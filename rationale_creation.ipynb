```python
responses={} 
cant_jsonify = []
import json
for i in range(len(text)):
    len_category_names = len(text[i].split(" ", 1)[1].strip().split(","))
    if len_category_names > 1:
        category = " also called ".join(category.strip() for category in text[i].split(" ", 1)[1].strip().split(","))
        name = text[i].split(" ", 1)[1].strip().split(",")[0]
    else:
        category = text[i].split(" ", 1)[1].strip().split(",")[0]
        name = category
    prompt_complete = prompt.format(category_name=category)
    full_prompt = json_part + prompt_complete
    response = model.invoke(full_prompt)
    print(f"Category: {category}")
    print(f"Response: {response}")
    cleaned_response = response.content.strip().strip('`').strip('json\n').strip().replace('\n', ' ')

    try:
        json.loads(cleaned_response)
        responses[name] = cleaned_response
    except json.JSONDecodeError as r:
        print(f"Error decoding JSON for category: {category}, Error: {r}")
        cant_jsonify.append(category)
        new_prompt = "fix the response to be a valid JSON format, heres the response and error look out for ommission on apostrophe for the python strings :" + f"\n{cleaned_response}\nError: {r} \n\n No other explanations, only provide the response"
        response2= model.invoke(new_prompt)
        cleaned_response2 = response2.content.strip().strip('`').strip('json\n').strip().replace('\n', ' ')

        if json.loads(cleaned_response):
            responses[name] = cleaned_response2

        else:
            cant_jsonify.append(category)
            print(f"No 'json-fiable' response for category: {category}")
    
```
